```python
numerical_columns = ['tenure', 'MonthlyCharges', 'TotalCharges']
data[numerical_columns] = data[numerical_columns].apply(pd.to_numeric, errors='coerce')
```

```python
for col in numerical_columns:
    plt.figure(figsize=(8, 6))
    data.boxplot(column=col)
    plt.title(f'Box Plot for {col}')
    plt.show()
```

```python
# 7. Identify and handle outliers using IQR method
# Calculate Q1, Q3, and IQR for each numerical column
Q1 = data[numerical_columns].quantile(0.25)
Q3 = data[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap the outliers to the boundary values
for col in numerical_columns:
    data[col] = np.where(data[col] < lower_bound[col], lower_bound[col], data[col])
    data[col] = np.where(data[col] > upper_bound[col], upper_bound[col], data[col])

# Display the modified data with outliers handled
print(data[numerical_columns].head())
```
