# Title : Analyzing Sales Performance by Region in a Retail Company

```python
#pr6
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./datasets/customer_shopping_data.csv") 
df.head()
```

```python
df.tail()
df['Total Amount'] = df['quantity'] * df['price'] 
```

```python
total_sales_by_region = df.groupby('shopping_mall')['Total Amount'].sum()
total_sales_by_region
```

```python
df
```

```python
total_sales_by_region = df.groupby('shopping_mall')['Total Amount'].sum()
total_sales_by_region
```

```python
df.groupby("shopping_mall").size()
```

```python
df.groupby("shopping_mall")['quantity'].count()
```

```python
# To check the count of records grouped by the product categories
df.groupby("category").count()
```

```python
# total sales for each mall branch
branch_sales = df.groupby("shopping_mall").agg({'Total Amount': 'sum'})
branch_sales
```

```python
# total sales for each category of product
category_sales = df.groupby("category").sum()
category_sales = df.groupby("category").agg({'Total Amount': 'sum'})
category_sales
```

```python
# to get the top performing branches
branch_sales.sort_values(by = "Total Amount", ascending = False)
```

```python
# to get the top selling categories
category_sales.sort_values(by = "Total Amount", ascending = False)
```

```python
# to get total sales for each combination of branch and product_category
combined_branch_category_sales = df.groupby(["shopping_mall", "category"]).sum()
combined_branch_category_sales
```

```python
# to get total sales for each combination of branch and product_category
combined_branch_category_sales = df.groupby(["shopping_mall", "category"]).sum()
combined_branch_category_sales
```

```python
# pie chart for sales by branch
plt.pie(branch_sales["Total Amount"], labels = branch_sales.index) 
plt.show()
branch_sales.index
```

```python
# pie chart for sales by product category
plt.pie(category_sales["Total Amount"], labels = category_sales.index) 
plt.show()
```

```python
combined_pivot = df.pivot_table(index="shopping_mall", columns="category", values="Total Amount", aggfunc="sum") 
```

```python
# grouped bar chart for sales of different categories at different branches
combined_pivot.plot(kind="bar", figsize=(10, 6)) 
plt.show()
```
